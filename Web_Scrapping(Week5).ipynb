```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url= 'http://books.toscrape.com/'
response= requests.get(url)
response.status_code
```

```python
soup= bs(response.text,'html')
type(soup)
```

```python
books_tag= soup.find_all('article',class_='product_pod')
len(books_tag)
```

Extracting books name,rating,price and link from the website

```python
# empty lists
titles= []
ratings= []
prices= []
links= []

for book_tag in books_tag:
  # Extract title of book
  title_tag= book_tag.find('a',title=True)
  title= title_tag['title'] if title_tag else None
  titles.append(title)

  # Extract Rating of book
  rating_tag= book_tag.find('p',class_='star-rating')
  rating= rating_tag['class'][1] if rating_tag else None
  ratings.append(rating)

  # Extract price of book
  price_tag= book_tag.find('p','price_color')
  price= price_tag.text if price_tag else None
  prices.append(price)

  # Extract link of the book
  link_tag= book_tag.find('a')['href']
  link= link_tag if link_tag else None
  links.append('http://books.toscrape.com/'+link)
```

Adding data extracted from website into dataframe

```python
data= pd.DataFrame({'Book_name':titles,'Rating':ratings,'Price':prices,'Link':links})
pd.set_option('display.max_colwidth', None)
data.head(10)
```
